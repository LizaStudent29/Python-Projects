Требуется параллельно (насколько возможно с помощью потоков) выполнить вычисления по формуле 1, вычисления по формуле 2. Затем выполнить вычисления по формуле 3, которые выглядят следующим образом: результат вычислений 1 + результат вычислений 2.

Выполнить последовательно на 10 000 итераций и 100 000 итераций.
Формула 1: f(x) = x^2-x^2+x4-x5+x+x
Формула 2: f(x) = x+x
Вывести длительность выполнения всех 10 000 итераций и 100 000 итераций в секундах в разбивке по шагам вычислений 1, 2 и 3.

```python
import time
import threading

def formula1(x):
    return x**2 - x**2 + x**4 - x**5 + x + x

def formula2(x):
    return x + x

def calculate_formula1(iterations):
    start_time = time.time()
    result = 0
    for _ in range(iterations):
        result += formula1(_)
    duration = time.time() - start_time
    return result, duration

def calculate_formula2(iterations):
    start_time = time.time()
    result = 0
    for _ in range(iterations):
        result += formula2(_)
    duration = time.time() - start_time
    return result, duration

def main():
    iterations = [10000, 100000]
    
    for it in iterations:
        print(f"\nNumber of iterations: {it}")
        thread1 = threading.Thread(target=calculate_formula1, args=(it,))
        thread2 = threading.Thread(target=calculate_formula2, args=(it,))
        
        start_time = time.time()
        thread1.start()
        thread2.start()
        thread1.join()
        thread2.join()
        duration_threads_creation = time.time() - start_time

        start_time = time.time()
        result1, duration1 = calculate_formula1(it)
        result2, duration2 = calculate_formula2(it)
        duration_sequential = time.time() - start_time

        result3 = result1 + result2

        print(f"Formula 1 result: {result1}, Duration: {duration1} seconds")
        print(f"Formula 2 result: {result2}, Duration: {duration2} seconds")
        print(f"Formula 3 result: {result3}, Duration: {duration_sequential} seconds")
        print(f"Threads creation duration: {duration_threads_creation} seconds")

if __name__ == "__main__":
    main()
```
