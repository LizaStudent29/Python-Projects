```python
import re

def validate_car_number(car_id):
    # Регулярное выражение для проверки формата номера
    pattern = re.compile(r'^[АВЕКМНОРСТУХABEKMHOPCTYX]{1}\d{3}[АВЕКМНОРСТУХABEKMHOPCTYX]{2}\d{2,3}$')

    # Проверяем соответствие номера формату
    if not pattern.match(car_id):
        return "Номер не валиден."

    # Извлекаем номер и регион
    number = car_id[:6]
    region = car_id[6:]

    return f"Номер {number} валиден. Регион: {region}."

# Примеры использования
car_id1 = 'А222ВС96'
result1 = validate_car_number(car_id1)
print(result1)

car_id2 = 'АБ22ВВ193'
result2 = validate_car_number(car_id2)
print(result2)
```
