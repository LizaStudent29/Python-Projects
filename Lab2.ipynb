```python
goods = input("Введите список товаров (разделите запятыми): ").split(",")

print("\n-------\n")

prices = {}

print(f"Введите названия магазинов (разделите запятыми):")
shops = input().split(",")

print("\n-------\n")

for ware in goods:
    prices[ware] = {}  
    
    for shop in shops:
        price = float(input(f"Цена '{ware}' в '{shop}': "))
        prices[ware][shop] = price

print("\n-------\n")

print("Введенные данные:")
for ware, prices_in_shops in prices.items():
    print(f"{ware}: {prices_in_shops}")
    
```

```python
total_price_dict = {}
for shop in shops:
    total_price = sum(prices[ware][shop] for ware in goods)
    total_price_dict[shop] = total_price

the_most_profitable_shop = min(total_price_dict, key=total_price_dict.get)

print("\nОбщая стоимость покупок в разных магазинах:")
for shop, temp_price in total_price_dict.items():
    print(f"{shop}: {temp_price} руб.")

print(f"\nВ {the_most_profitable_shop} вы можете сэкономить больше всего денег.")
```
