Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}
Пример работы программы при выводе первых двух элементов словаря purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘

```python
import random
import json

# Генерация случайных данных для примера
categories = ['Продукты', 'Электроника', 'Одежда', 'Косметика']
user_ids = ['user' + str(i) for i in range(1, 11)]

data = [{'user_id': random.choice(user_ids), 'category': random.choice(categories)} for _ in range(20)]

# Запись данных в файл purchase_log.txt
with open('purchase_log.txt', 'w', encoding='utf-8') as f:
    f.write('user_id,category\n')  # Заголовок файла
    for entry in data:
        f.write(f"{entry['user_id']},{entry['category']}\n")
```

```python
# Чтение содержимого файла purchase_log.txt и создание словаря purchases
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    next(f)  
    purchases = {line.strip().split(',')[0]: line.strip().split(',')[1] for line in f}

count = 0
for key, value in purchases.items():
    print(key, value)
    count += 1
    if count == 2:
        break
```
