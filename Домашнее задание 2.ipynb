Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечётное;
две средних буквы, если число букв чётное.
Пример работы программы:

word = 'test’
Результат: es
word = 'testing’
Результат: t

```python
def middle_letters(word):
    length = len(word)
    if length % 2 == 1:  # Если количество букв нечетное
        middle_index = length // 2
        result = word[middle_index]
    else:  # Если количество букв четное
        middle_index = length // 2
        result = word[middle_index - 1 : middle_index + 1]
    return result

word1 = 'test'
result1 = middle_letters(word1)
print(f"Для слова '{word1}' результат: {result1}")

word2 = 'testing'
result2 = middle_letters(word2)
print(f"Для слова '{word2}' результат: {result2}")
```

Напишите код на Python в среде Jupyter Notebook для решения следующей задачи.

Вы делаете MVP (минимально жизнеспособный продукт) dating-сервиса.
У вас есть список юношей и девушек.
Выдвигаем гипотезу: лучшие рекомендации получатся, если просто отсортировать имена по алфавиту и познакомить людей с одинаковыми индексами после сортировки. Но вы не будете никого знакомить, если кто-то может остаться без пары.

Примеры работы программы:

boys = [‘Peter’, ‘Alex’, ‘John’, ‘Arthur’, ‘Richard’]
girls = [‘Kate’, ‘Liza’, ‘Kira’, ‘Emma’, ‘Trisha’]

Результат
Идеальные пары:
Alex и Emma
Arthur и Kate
John и Kira
Peter и Liza
Richard и Trisha

boys = [‘Peter’, ‘Alex’, ‘John’, ‘Arthur’, ‘Richard’, ‘Michael’]
girls = [‘Kate’, ‘Liza’, ‘Kira’, ‘Emma’, ‘Trisha’]

Результат: Внимание, кто-то может остаться без пары.

```python
def find_perfect_couples(boys, girls):
    if len(boys) != len(girls):
        print("Внимание, кто-то может остаться без пары.")
        return

    sorted_boys = sorted(boys)
    sorted_girls = sorted(girls)

    perfect_couples = zip(sorted_boys, sorted_girls)

    print("Идеальные пары:")
    for couple in perfect_couples:
        print(f"{couple[0]} и {couple[1]}")

boys1 = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls1 = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

find_perfect_couples(boys1, girls1)

boys2 = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
girls2 = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

find_perfect_couples(boys2, girls2)
```
